# Inverse encoding / Inverse Transform / Decoding categorical variables

Inverse encoding is the process of converting the encoded data back to its origical form. This is done by using the ```inverse_transform()``` method. this method is available for the following encoders:

- `LabelEncoder`
- `OrdinalEncoder`
- `OneHotEncoder`

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
````

````python
# load the data 
df = sns.load_dataset("titanic")
df.head()
````

````python
# impute missin values for age, embarked , embark_town and dec
df['age'].fillna(df['age'].mean(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)

# drop deck colum
df.drop('deck', axis=1, inplace=True )
````

````python
df.info()
````

````python
encode = OneHotEncoder()

df['sex'] = encode.fit_transform(df[['sex']]).toarray()
df.head()
````

````python
# encode all categorical variables and object variables

le = LabelEncoder()

for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'category':
        df[col] = le.fit_transform(df[col])
````

````python
df.info()
````

````python
df.head()
````

````python
df['sex'] = le.inverse_transform(df['sex'])
df.head()
````
