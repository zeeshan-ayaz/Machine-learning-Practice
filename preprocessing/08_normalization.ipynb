# log transformation

```python
import pandas as pd
import numpy as np
```

```python
# creat a data
data = {'value':[10,20,50,100,200,500,1000,4000,20000,50000]}
data = pd.DataFrame(data)
data
```

```python
data['log_values'] = np.log(data['value'])
data
```

```python
import seaborn as sns
# sns.histplot(data['value'])
sns.histplot(data['log_values'], kde=True)
```

```python
import matplotlib.pyplot as plt
```

```python
for col in data.columns:
    sns.histplot(data[col], kde=True)
    plt.show()
```
