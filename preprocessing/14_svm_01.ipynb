# Support Vector Machines

support vector machines (svm) is `supervised machine learning algorithm` which can be used for both classification or regression challenges.

However, it is mostly used in classificatin problems.

In this algorithm, we plot each data item as a point in n-dimensional space (where n is number of features you have) with the value of each feature being the value of a particular coordinate. Then, we perform classification by finding the hyper-plane that differentiate the two classes very well.

**Types of Kernels:**
1. Linear kernel
2. Polynomial kernel
3. Radial Basis Function kernel (RBF)


```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ML libraries
from sklearn.svm import LinearSVC, NuSVC, SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
df['species'].value_counts()
```

```python
# lets make x and y in our data
x = df.drop('species', axis=1) #without this colums adopt all colums 
y = df['species']
```

```python
# train test split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# call the model
model = SVC()
```

```python
# fit the model
model.fit(x_train, y_train)
```

```python
#predict the model
y_pred = model.predict(x_test)
```

```python
# evaluate the model
print("confusion matrix: ", confusion_matrix(y_test, y_pred))
print("classification Report: \n", classification_report(y_test, y_pred))
```

```python
# draw the confusion matrix in ploty
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True)
plt.xlabel('predicted')
plt.ylabel("actual")
plt.show()
```

```python
# train split data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
# call the model
model = SVR()
```

```python
# fit the model
model.fit(x_train, y_train)
```

```python
# prediction
y_pred = model.predict(x_test)
y_pred
```
